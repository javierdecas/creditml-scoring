# Análisis Exploratorio de Datos (EDA) - Credit Scoring

Este notebook realiza un análisis exploratorio básico del dataset de aprobación de préstamos. Incluye visualizaciones de la variable objetivo y análisis de correlaciones.

**Dataset:** credit_approval.csv  
**Variable objetivo:** approved (0 = rechazado, 1 = aprobado)

```python
# Importar librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8')

# Configurar gráficos
sns.set_palette("husl")
```

```python
# Cargar datos
df = pd.read_csv('../data/raw/credit_approval.csv')
print(f"Dataset shape: {df.shape}")
print("\nPrimeras filas:")
df.head()
```

```python
# Información básica del dataset
print("Información del dataset:")
df.info()

print("\nEstadísticas descriptivas:")
df.describe()
```

```python
# Verificar valores nulos
print("Valores nulos por columna:")
df.isnull().sum()
```

## Análisis de la Variable Objetivo

```python
# Distribución de la variable objetivo
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=df, x='approved', palette='Set2')
plt.title('Distribución de la Variable Objetivo (Approved)', fontsize=14)
plt.xlabel('Approved (0 = Rechazado, 1 = Aprobado)')
plt.ylabel('Número de Solicitudes')

# Agregar porcentajes
total = len(df)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom')

plt.show()

print(f"Proporción de aprobados: {df['approved'].mean():.3f}")
```

## Análisis de Variables Numéricas

```python
# Variables numéricas
numeric_cols = ['age', 'income', 'loan_amount', 'credit_history_length', 'existing_debt']

# Histogramas
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    sns.histplot(data=df, x=col, ax=axes[i], kde=True)
    axes[i].set_title(f'Distribución de {col}', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Boxplots por variable objetivo
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    sns.boxplot(data=df, x='approved', y=col, ax=axes[i])
    axes[i].set_title(f'{col} vs Approved', fontsize=12)

plt.tight_layout()
plt.show()
```

## Análisis de Variables Categóricas

```python
# Variables categóricas
categorical_cols = ['employment_status', 'payment_history']

# Countplots
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

for i, col in enumerate(categorical_cols):
    sns.countplot(data=df, x=col, ax=axes[i], palette='Set3')
    axes[i].set_title(f'Distribución de {col}', fontsize=12)
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Variables categóricas vs objetivo
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

for i, col in enumerate(categorical_cols):
    # Calcular porcentajes
    cross_tab = pd.crosstab(df[col], df['approved'], normalize='index')
    cross_tab.plot(kind='bar', stacked=True, ax=axes[i], colormap='Set2')
    axes[i].set_title(f'{col} vs Approved (porcentajes)', fontsize=12)
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].legend(title='Approved', labels=['Rechazado', 'Aprobado'])

plt.tight_layout()
plt.show()
```

## Análisis de Correlaciones

```python
# Codificar variables categóricas para correlación
df_encoded = df.copy()
df_encoded['employment_status'] = df_encoded['employment_status'].astype('category').cat.codes
df_encoded['payment_history'] = df_encoded['payment_history'].astype('category').cat.codes

# Matriz de correlación
correlation_matrix = df_encoded.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Matriz de Correlación', fontsize=14)
plt.show()
```

```python
# Correlación con la variable objetivo
target_corr = correlation_matrix['approved'].sort_values(ascending=False)
print("Correlaciones con la variable objetivo (approved):")
print(target_corr)
```
